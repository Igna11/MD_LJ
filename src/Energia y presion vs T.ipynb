```python
import numpy as np
import os
import glob
import matplotlib.pyplot as plt
```

```python
Dir = r"C:\Users\Paula\Documents\Github\MD_LJ\src"
os.chdir(Dir)
```

```python
filenames_Epot = glob.glob('Epot_vs_T_rho*.txt')
filenames_Ecin = glob.glob('Ecin_vs_T_rho*.txt')
filenames_P = glob.glob('P_vs_T_rho*.txt')

for i in range (len(filenames_Epot)):
    data_Epot = np.loadtxt(fname=filenames_Epot[i])
    data_Ecin = np.loadtxt(fname=filenames_Ecin[i])
    data_P = np.loadtxt(fname=filenames_P[i])

    T = data_Epot[:,0]
    Epot = []
    Ecin = []
    Etot = []
    P = []
    
    for j in range(len(data_Epot[:,0])):
        Epot.append(np.mean(data_Epot[j,:]))
   
    for j in range(len(data_Ecin[:,0])):
        Ecin.append(np.mean(data_Ecin[j,:]))
    
    for j in range(len(data_Epot[:,0])):
        Etot.append(np.mean(data_Epot[j,:])+np.mean(data_Ecin[j,:]))
    
    for j in range(len(data_P[:,0])):
        P.append(np.mean(data_P[j,:]))    
        
        
    plt.figure(figsize=(9,7))  
#     plt.plot(T,Epot,'.',label= 'Energía potencial')
#     plt.plot(T,Ecin,'.',label= 'Energía cinética')
#     plt.plot(T,Etot,'.',label = 'Energía total')
    plt.plot(T,P,'.',label= 'Presión')
    plt.grid(True);
    plt.xlabel('Temperatura', fontsize = 15);
    plt.legend(fontsize = 15);
```

```python
plt.plot(T,Epot,'.',label = 'pot')
plt.plot(T,Ecin,'.', label = 'cin')
plt.plot(T,P,'.', label = 'p')
plt.legend()
```
