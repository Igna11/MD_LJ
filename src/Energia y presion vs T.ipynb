```python
# Para graficar
valid_markers = ([item[0] for item in mpl.markers.MarkerStyle.markers.items() if
item[1] is not 'nothing' and not item[1].startswith('tick') and not item[1].startswith('caret')])
```

```python
filenames_P = glob.glob('P_vs_T_rho*.txt')

plt.figure(figsize=(9,7))  
for i in range (len(filenames_P)):
    string = filenames_P[i]
    data_P = np.loadtxt(fname=filenames_P[i])

    T = data_P[:,0]
    P = []

    for j in range(len(data_P[:,0])):
        P.append(np.mean(data_P[j,:]))    
        
    plt.plot(T,P,marker = np.random.choice(valid_markers),linewidth = 0, markersize = 10,label= r'Presi√≥n para $\rho$ %s'%string[10:-8])
    plt.grid(True);
    plt.xlabel('Temperatura', fontsize = 15);
    plt.legend(fontsize = 15);
    plt.xticks(fontsize = 15);
    plt.yticks(fontsize = 15);
    plt.savefig('PresionvsT.pdf')
```

```python
filenames_Epot = glob.glob('Epot_vs_T_rho*.txt')
filenames_Ecin = glob.glob('Ecin_vs_T_rho*.txt')

plt.figure(figsize=(9,7))  
for i in range (len(filenames_Epot)):
    string = filenames_Epot[i]
    data_Epot = np.loadtxt(fname=filenames_Epot[i])
    data_Ecin = np.loadtxt(fname=filenames_Ecin[i])
    
    T = data_Epot[:,0]
    Etot = []
    
    for j in range(len(data_Epot[:,0])):
        Etot.append(np.mean(data_Epot[j,:])+np.mean(data_Ecin[j,:]))
        
    plt.plot(T,Etot,marker = np.random.choice(valid_markers),linewidth = 0, markersize = 10,label= r'Energ√≠a total para $\rho$ %s'%string[13:-8])
    plt.grid(True);
    plt.xlabel('Temperatura', fontsize = 15);
    plt.legend(fontsize = 15);
    plt.xticks(fontsize = 15);
    plt.yticks(fontsize = 15);
    #plt.savefig('EnergiavsT.pdf')
```
